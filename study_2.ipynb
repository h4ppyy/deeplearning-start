```python
import matplotlib.pyplot as plt
```

### 1차함수를 코드로 구현하여 matplotlib에 출력하기

$$H(x) = x$$

$$H(x) = 2x$$

```python
def linear(input, w, y):
    output = []
    for x in input:
        h = w * x + y
        output.append(h)
    return output
```

```python
x = [0, 1, 2, 3, 4] # input
y = linear(x, 1, 0) # solution
y_hat = linear(x, 2, 0) # predict

plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y)
plt.plot(x, y_hat)
plt.grid(True)
plt.show()
```

### cost function(MSE)를 코드로 구현하여 matplotlib에 출력하기

$$cost(w) = \frac{1}{m} \sum_{i=1}^{m} (wx_i-y_i)^2 $$

```python
# MSE
def cost(x, y, w):
    output = []
    for w in w:
        sum = 0
        for n in range(0, len(x)):
            sum += ((w * x[n]) - y[n])**2
        cost_w = sum/len(x)
        output.append(cost_w)
    return output
```

```python
x = [0, 1, 2]
y = [0, 1, 2]
w = [-3, -2, -1, 0, 1, 2, 3, 4, 5]
cost_w = cost(x, y, w)
plt.xlabel('w')
plt.ylabel('cost(w)')
plt.plot(w, cost_w)
plt.grid(True)
plt.show()
```
